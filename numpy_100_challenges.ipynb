```python
# Challenge 1: Import NumPy and print its version
print("Numpy Version:",np.__version__)

# Challenge 2: Create a 1D array of numbers from 0 to 9
arr=np.arange(10)
print(arr)

# Challenge 3: Create a 3x3 NumPy array filled with True
arr2 = np.full((3, 3), True, dtype=bool)
print("\nChallenge 3:\n", arr2)

# Challenge 4: Extract all odd numbers from arr1
odd_numbers=arr[arr%2==1]
print("Odd number from arr\n:", odd_numbers)

# Challenge 5: Replace all odd numbers in arr1 with -1
arr1_mod=np.where(arr%2==1 ,-1,arr)
print("\nChallenge 5 (odds to -1):",arr1_mod)

# Challenge 6: Reshape a 1D array of 10 elements into 2 rows
reshaped_arr = arr.reshape(2, 5)
print("Reshaped array:\n",reshaped_arr)

# Challenge 7: Stack two arrays vertically
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
vstacked = np.vstack((a, b))
print("\nChallenge 7:\n", vstacked)

# Challenge 8: Stack the same arrays horizontally
hstacked = np.hstack((a, b))
print("\nChallenge 8:\n", hstacked)

# Challenge 9: Create a 5x5 matrix with values 1,2,3,4 just below the diagonal
mat = np.diag([1, 2, 3, 4], k=-1)
print("\nChallenge 9:\n", mat)

# Challenge 10: Create a checkerboard 8x8 matrix (0s and 1s)
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1
checkerboard[::2, 1::2] = 1
print("\nChallenge 10 (Checkerboard):\n", checkerboard)

```

```python
#Challenge 71-80
# Challenge 71: Find common values between two arrays
a = np.array([1, 2, 3, 4, 5])
b = np.array([4, 5, 6, 7, 8])
common=np.intersect1d(a,b)
print("\nChallenge 71:\nCommon Elements:", common)

# Challenge 72: Find differences between two arrays
diff = np.setdiff1d(a, b)
print("\nChallenge 72:\nElements in 'a' not in 'b':", diff)

# Challenge 73: Get the union of two arrays
union = np.union1d(a, b)
print("\nChallenge 73:\nUnion:", union)

# Challenge 75: Normalize an array (0 to 1 range)
arr = np.random.randint(1, 100, 10)
normalized_arr=(arr-arr.min())/(arr.max()-arr.min())
print("\nChallenge 75:\nNormalized Array:", normalized_arr)

# Challenge 76: One-hot encode a NumPy array
arr = np.array([0, 1, 2])
one_hot = np.eye(3)[arr]
print("\nChallenge 76:\nOne-hot Encoded:\n", one_hot)

# Challenge 77: Convert boolean array to int
bool_arr = np.array([True, False, True])
int_arr = bool_arr.astype(int)
print("\nChallenge 77:\nConverted to Int:", int_arr)

# Challenge 78: Multiply a matrix by a scalar
mat = np.array([[1, 2], [3, 4]])
scaled = mat * 10
print("\nChallenge 78:\nMatrix Scaled by 10:\n", scaled)

# Challenge 79: Transpose a matrix
transposed = mat.T
print("\nChallenge 79:\nTransposed Matrix:\n", transposed)

# Challenge 80: Flatten a multi-dimensional array
arr = np.array([[1, 2], [3, 4], [5, 6]])
flat = arr.flatten()
print("\nChallenge 80:\nFlattened Array:", flat)
```
